## Problem 1: Even Number Skipper

```python
# Write a program that displays numbers from 1 to 20, but skips all even numbers.
print("Numbers from 1 to 20, skipping even numbers:")
for i in range(1, 21):
    if i % 2 == 0:  # Check if the number is even
        continue  # Skip to the next iteration if even
    print(i)
```

## Problem 2: Password Checker

```python
# Write a program that keeps asking the user to enter a password until the correct password is entered.

correct_password = "ColabRocks"
attempts = 0
max_attempts = 3

print("Password Checker")
while True:
    if attempts >= max_attempts:
        print("Too many incorrect attempts. Access denied.")
        break

    password_guess = input("Enter the password: ")
    attempts += 1

    if password_guess == correct_password:
        print("Access granted! Welcome.")
        break  # Exit the loop if password is correct
    else:
        print(f"Incorrect password. You have {max_attempts - attempts} attempts left.")
```

## Problem 3: Number Filter Display

```python
# Write a program that displays numbers from 1 to 10, but:
# Skips number 4
# Does nothing special for number 6
# Stops completely when the number reaches 8

print("Numbers from 1 to 10 with specific filters:")
for i in range(1, 11):
    if i == 4:
        continue  # Skip number 4
    elif i == 6:
        pass  # Do nothing special for number 6
    elif i == 8:
        break  # Stop the loop when number 8 is reached
    print(i)
```
