```python
score = input("How many scores will you input?")
```

```python
num_scores = int(score)
scores = []
total_score = 0

for i in range(num_scores):
    while True:
        try:
            current_score = int(input(f"Enter score: "))
            scores.append(current_score)
            total_score += current_score
            break
        except ValueError:
            print("Invalid input. Please enter a number.")

average_score = total_score / num_scores

print(f"The total score is: {total_score}")
print(f"The average is: {average_score:.1f}")
print("PASSED")
```
